```python
import glob
import pandas as pd 
```

```python
# Specify what folders do we need
INPUT_FOLDER      = './data_preprocessed3/'
COLUMNS = ['velocity', 'acceleration', 'trajectory_id', 'subfolder',
        'labels', 'v_ave', 'v_med', 'v_max', 'v_std', 'a_ave', 'a_med', 'a_max', 'a_std']
```

```python
list_df_metadata = []

# Find all files that ends with .csv, read and append to the list_df_metadata
for filename in glob.glob(INPUT_FOLDER + '*.csv'):
    df_data = pd.read_csv(filename)
    list_df_metadata.append(df_data)
    
df_metadata = pd.concat(list_df_metadata)
df_metadata = df_metadata.drop(['Unnamed: 0'], axis=1)
df_metadata = df_metadata.reset_index(drop=True)
df_metadata.head()
```

# Dim reduction algorithms

```python
from sklearn.decomposition import TruncatedSVD

def PCA2(X,y):
    """PCA"""
    PCA_model = TruncatedSVD(n_components=2)
    reduced = PCA_model.fit_transform(X)
    return reduced,y

from sklearn.decomposition import PCA

def NPCA(X, y):
    pca = PCA(n_components=2)
    res = pca.fit_transform(X)
    return res, y

import numpy as np
from sklearn.manifold import TSNE
def tSNE(X,y,perp=10):
    """Subsample to prevent memory error
    - perp: the perplexity for the tSNE algo"""

    tSNE_model = TSNE(verbose=2,perplexity=perp)
    reduced_tsne = tSNE_model.fit_transform(X)
    return reduced_tsne,y
```

```python
X_COLUMNS = ['v_rol', 'v_rsd', 'v_qu1','v_qu2', 'v_qu3', 'a_rol', 'a_rsd', 'a_qu1', 'a_qu2', 'a_qu3']
Y_COLUMN  = ['labels']
```

```python
df_metadata = df_metadata.dropna()
X = df_metadata[X_COLUMNS]
Y = df_metadata[Y_COLUMN]
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
red, y =  NPCA(X, Y)
le = preprocessing.LabelEncoder()
colors = le.fit_transform(Y['labels'])
plt.scatter(red[:, 0], red[:, 1],c=colors, marker='o', linewidths=0)
plt.show()
```

```python
red, y =  PCA2(X, Y)
le = preprocessing.LabelEncoder()
colors = le.fit_transform(Y['labels'])
plt.scatter(red[:, 0], red[:, 1],c=colors, marker='o', linewidths=0)
plt.show()
```

```python
red, y =  tSNE(X, Y)
le = preprocessing.LabelEncoder()
colors = le.fit_transform(Y['labels'])
plt.scatter(red[:, 0], red[:, 1],c=colors, marker='o', linewidths=0)
plt.show()
```

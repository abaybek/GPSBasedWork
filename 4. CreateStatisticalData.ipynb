```python
import pandas as pd
import glob
import gc
```

```python
INPUT_FOLDER      = './data_preprocessed/'
IMPORTING_COLUMNS = ['velocity', 'acceleration', 'trajectory_id', 'subfolder',
                     'labels', 'v_ave', 'v_med', 'v_max', 'a_ave', 'a_med', 'a_max',
                     'datetime', 'timedelta']
```

```python
import datetime as dt
list_df_metadata = []

# Convert str datetime to datetime format
def to_datetime(string):
    return dt.datetime.strptime(string, '%Y-%m-%d %H:%M:%S')

def grouping_data(data):
    data['year'] = data['datetime'].apply(lambda x: x.year)
    data['month'] = data['datetime'].apply(lambda x: x.month)
    data['day'] = data['datetime'].apply(lambda x: x.day)
    
    g=data.groupby(['year', 'month', 'day'])['velocity', 'acceleration', 'v_ave', 'v_med','v_med'].mean()
    print(g)

for filename in glob.glob(INPUT_FOLDER + '*.csv'):
    df_data = pd.read_csv(filename, usecols=IMPORTING_COLUMNS, 
                          date_parser=to_datetime,
                          parse_dates=['datetime'])
    grouping_data(df_data)
    list_df_metadata.append(df_data)
    break
    
df_metadata = pd.concat(list_df_metadata)
df_metadata.head()
```
